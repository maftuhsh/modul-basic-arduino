## Serial monitor
 Serial Monitor adalah fitur di Arduino IDE (dan IDE mikrokontroler lain) yang dipakai untuk melihat dan mengirim data lewat komunikasi serial (UART) antara komputer dan mikrokontroler (Arduino, ESP32, ESP8266, dll). 

 atau simpelnya Serial Monitor adalah jendela teks di Arduino IDE yang dipakai untuk komunikasi antara Arduino dan komputer.

Bahasa gampangnya:
     Serial Monitor = layar chat antara Arduino dan PC

Fungsinya Buat Apa?
Serial Monitor dipakai untuk:
    - Melihat pesan dari Arduino
    - Mengirim perintah teks ke Arduino
    - Mengecek apakah program berjalan atau tidak
    - Debug (nyari kesalahan program)

Cara Kerja Singkatnya
Arduino dan komputer saling kirim data lewat kabel USB
Data dikirim satu per satu (serial, berurutan).

```python
void setup() {
  Serial.begin(9600); // menampilkan di serial monitor
  Serial.println("Hello World");
  Serial.println(("DAY 1 belajar arduino"));
}

void loop() {
}
```
