```python
## Apa itu void?
Dalam pemrograman Arduino, setiap program terdiri dari fungsi.
Fungsi adalah sekumpulan perintah yang diberi nama. 

Kata void dipakai untuk menandakan bahwa:
- Fungsi tersebut hanya menjalankan perintah, tidak menghasilkan nilai atau jawaban apa pun.

Artinya, setelah fungsi dijalankan:
- Tidak ada angka
- Tidak ada teks
- Tidak ada hasil yang dikembalikan

Contoh sederhana:
- Menyalakan LED
- Mematikan motor
- Mengatur pin

Semua itu aksi, bukan perhitungan, jadi memakai void.
```

```python
## Fungsi void setup()

setup() adalah fungsi khusus yang pasti ada di setiap program Arduino.

Kapan setup() dijalankan?
- Dijalankan satu kali saja
Saat:
    - Arduino pertama kali dinyalakan
    - Arduino di-reset
    - Program baru di-upload

Untuk apa setup()?
Digunakan untuk persiapan awal, sebelum Arduino mulai bekerja terus-menerus.

Biasanya berisi:
    - Mengatur pin sebagai INPUT atau OUTPUT
    - Memulai komunikasi serial
    - Menginisialisasi sensor atau modul
```

```python
void setup() {
  // fungsi yang di kerjakan sekali saat arduino menyala (program yang berjalan selama kali)
  pinMode(13, OUTPUT); // Pin 13 disiapkan sebagai output (misalnya untuk LED)
  serial.begin(9600) // Arduino siap mengirim dan menerima data lewat Serial Monitor
}

// Semua ini cukup dilakukan sekali, tidak perlu diulang.
```

## Fungsi void loop()
Setelah setup() selesai, Arduino akan masuk ke fungsi loop(). `loop()` akan terus dijalankan secara **berulang selama Arduino menyala**.

Bagaimana cara kerja loop()?
Isi loop() akan:
    - Dijalanin dari atas ke bawah
    - Setelah selesai, diulang lagi dari awal
    - Berjalan terus selama Arduino menyala
Arduino tidak pernah berhenti menjalankan loop().

Untuk apa loop()?
Digunakan untuk:
    - Membaca sensor
    - Menyalakan atau mematikan LED
    - Mengontrol motor
    - Menjalankan logika program utama

```python
void loop() {
  digitalWrite(13, HIGH); // LED dinyalakan
  delay(1000); // Tunggu 1 detik
  digitalWrite(13, LOW); // LED dimatikan
  delay(1000); // Tunggu 1 detik
} // Ulangi lagi dari awal

// Karena berada di loop(), proses ini akan berjalan terus menerus.
```
