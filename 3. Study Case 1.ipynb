```python
// jawaban 
int LED = 12;
char STATUS;

void setup(){

  pinMode(LED,HIGH);
  Serial.begin(9600); // menampilkan di serial monitor
  Serial.println("Hello World");
  Serial.println(("DAY 1 belajar arduino"));
}

void loop(){
  if (Serial.available() > 0){
    STATUS = Serial.read();
    Serial.println(STATUS);
    
  }
  // Kode ini berfungsi untuk mengecek apakah ada data masuk dari Serial Monitor, 
  // membaca satu karakter yang dikirim, lalu menampilkannya kembali ke Serial Monitor 
  // sebagai konfirmasi.
  if (STATUS == '1'){
    digitalWrite(LED,HIGH);
    Serial.println("LED MENYALA");
    delay(1000);
  }
  if (STATUS == '0'){
    digitalWrite(LED,LOW);
    Serial.println("LED MATI");
    delay(1000);
  }

}
```

```python
// jawaban 
int LED = 12;
String STATUS = "";

void setup(){

  pinMode(LED,HIGH);
  Serial.begin(9600); // menampilkan di serial monitor
  Serial.println("Hello World");
  Serial.println(("DAY 1 belajar arduino"));
}

void loop(){
  if (Serial.available() > 0){
    STATUS = Serial.readStringUntil('\n');
    STATUS.toUpperCase();
    Serial.println("PESAN DITERIMA");
    Serial.print(STATUS);
    
  }
  // Kode ini berfungsi untuk mengecek apakah ada data masuk dari Serial Monitor, 
  // membaca satu karakter yang dikirim, lalu menampilkannya kembali ke Serial Monitor 
  // sebagai konfirmasi.
  if (STATUS == "ON"){
    digitalWrite(LED,HIGH);
    Serial.println("LED MENYALA");
    delay(1000);
  }
  if (STATUS == "OFF"){
    digitalWrite(LED,LOW);
    Serial.println("LED MATI");
    delay(1000);
  }

}
```
